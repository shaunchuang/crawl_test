```python
import requests
import urllib
import pandas as pd
from pprint import pprint
```

```python

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36 Edg/88.0.705.68'}

df=pd.DataFrame(columns=['id','m_id','auto_brand_en','auto_brand_id','auto_gas_size','auto_make_date','auto_mileage_num','auto_price','auto_price_tag',
                         'auto_price_type','auto_title_all','item_post_date','item_renew_date','is_new','item_show_num','photo','photoList','rank'])

for i in range(3):
    url = f'https://auto.8891.com.tw/usedauto-newSearch.html?page={i}&limit=40'
    r=requests.get(url,headers=headers)
    data=r.json()
    df=df.append(pd.concat([df,pd.json_normalize(data['data']['data'])], axis=0))
    df.reset_index()
    df.to_csv('car_8891.csv', mode='a')

```
