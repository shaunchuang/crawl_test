```python
import requests
import pandas as pd
from pprint import pprint
from bs4 import BeautifulSoup
import time
import sys
```

```python
def get_html(url,headers):
    num_retry = 0  # 設定初始重試次數
    max_retry = 3  # 最大重試次數3次
    try:
        while num_retry <= max_retry:
            r = requests.get(url,headers=headers)
            if r.status_code == requests.codes.ok: # 是否成功請求網頁
                print('request success: %s' % url) 
                return r
            else:
                time.sleep(3)
                if num_retries == max_retries: # 當重試三次後丟出錯誤
                    raise exc
                print('retrying request')
                num_retries += 1
    except exc:
        print('http request error: %s' % url) # 超過請求次數印出 http request eror 並將錯誤丟出function外
        raise
    except:
        print('Unexcepted error: %s' % url, sys.exc_info()) # 出現意料之外的錯誤時印出錯誤並將錯誤丟出function外
        raise
```

```python
def parse_html(r):
    soup= BeautifulSoup(r.text,'lxml')
    if soup != None :
        print('parsing success: %s' % url) # 印出成功解析網頁名
        return soup
    else:
        raise exce  # 當beautifulsoup解析失敗為Nonetype時，將錯誤丟出function外
    
```
