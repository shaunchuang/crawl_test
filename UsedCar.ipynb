```python
#Second Step: Get the detail of the cars
df_detail_basic=pd.DataFrame(columns=['id', '車款', '車型', '年份', '里程','車輛價格', '縣市', '顏色','車門數','座位數', '排氣量','驅動方式','燃料','變速系統','刊登標題','照片網址','車商名','車商地址'])
# df_equipements=pd.DataFrame(columns=['皮椅','晶片系統','恆溫空調','VCD','ABS','CD','電動窗','中控鎖','電動後視鏡','天窗','霧燈','DVD','倒車顯影','衛星導航','TCS防滑','防盜','安全氣囊','定速','一手車','電動座椅','核桃木飾板','4WD','倒車雷達','行車電腦','總代理','MP3','柴油','ＨＩＤ頭燈','鋁圈'])

df1 = []
for j in websites: 
    url=f'https://used.carnews.com/UsedCar/Details/{j}' 
    r=requests.get(url,headers=headers)
    soup=BeautifulSoup(r.text,'lxml')


    # id=j
   
    # 車款
    car_brand1 = soup.find(class_='uc-breadcrumb')
    car_brand2 = car_brand1[0]
    car_brand3 = car_brand2.find_all('li')
    car_brand4 = car_brand3[2]
    car_brand = car_brand4.find('a').text.strip()
    #車型 
    
    car_brand5 = car_brand4 = car_brand3[3]
    car_model1 = car_brand5.find('a').text.strip()
    car_brand6 = car_brand4 = car_brand3[4]
    car_model2 = car_brand6.find('a').text.strip()
    car_model = car_model1 + ' ' + car_model2
    # 車身型式 car_body


    # 車輛年份
    detail_year = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_year1 = detail_year.find_all('tr')
    detail_year2 = detail_year1[0]
    detail_year3 = detail_year2.find_all('td')
    car_year = detail_year3[1].text.strip()
 
    # 里程
    detail_basic1 = soup.find(class_='pure-u-2-3 uc-basic-info')
    detail_basic2 = detail_basic1.find_all('td')
    car_mileage = detail_basic2[3].text.strip()
    
    # 車輛價格
    car_price = soup.find(class_='price').text.strip()

    # 車輛(商)所在地
    detail_city = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_city1 = detail_city.find_all('tr')
    detail_city2 = detail_city1[2]
    detail_city3 = detail_city2.find_all('td')
    car_address = detail_city3[1].text.strip() 
    
    # 車輛顏色
    detail_city = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_city1 = detail_city.find_all('tr')
    detail_city2 = detail_city1[1]
    detail_city3 = detail_city2.find_all('td')
    car_color = detail_city3[1].text.strip()    
    
    # 車門數
    detail_city = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_city1 = detail_city.find_all('tr')
    detail_city2 = detail_city1[3]
    detail_city3 = detail_city2.find_all('td')
    car_door = detail_city3[3].text.strip()
    
    # 座位數
    detail_city = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_city1 = detail_city.find_all('tr')
    detail_city2 = detail_city1[3]
    detail_city3 = detail_city2.find_all('td')
    car_seat = detail_city3[1].text.strip()
    # 排氣量
    detail_basic1 = soup.find(class_='pure-u-2-3 uc-basic-info')
    detail_basic3 = detail_basic1.find_all('tr')
    detail_basic4 = detail_basic3[1]
    detail_basic5 = detail_basic4.find_all('td')
    car_cylinderVolume = detail_basic5[3].text.strip()

    # 汽缸數(或形式)
    
    # 驅動方式
    detail_city = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_city1 = detail_city.find_all('tr')
    detail_city2 = detail_city1[5]
    detail_city3 = detail_city2.find_all('td')
    car_driveMode = detail_city3[1].text.strip()
    # 燃料
    detail_city = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_city1 = detail_city.find_all('tr')
    detail_city2 = detail_city1[4]
    detail_city3 = detail_city2.find_all('td')
    car_fuel = detail_city3[1].text.strip()
    # 變速系統
    detail_city = soup.find(class_='uc-car-detailspec uc-tab-content uc-card-container')
    detail_city1 = detail_city.find_all('tr')
    detail_city2 = detail_city1[4]
    detail_city3 = detail_city2.find_all('td')
    car_gear = detail_city3[3].text.strip()
    # 刊登標題
    car_title = soup.find('span',class_='uc-card-title').text.strip()
    # 照片網址
    car_img = soup.find(class_='uc-gallery-carousel')
    car_photo = car_img.find('img')['src']

    # 車商id
    
    # 車商名
    dealer_name = soup.find('h3',class_='uc-title').text.strip()
    # 賣家身份
    
    # 車商地址
    addr = soup.find_all('div',class_='uc-panel-content')
    addr1 = addr[1]
    addr2 = addr1.find_all('dl')
    addr3 = addr2[0]
    dealer_address = addr3.find('dd').text.strip()

    # 是否有認證

    
    df1.append([j,car_brand,car_model,car_year,car_mileage,car_price,car_address,car_color,car_door,car_seat,car_cylinderVolume,car_driveMode,car_fuel,car_gear,car_title,car_photo,dealer_name,dealer_address])
    df2 = pd.DataFrame(df1,columns=df_detail_basic.columns)

    
```
